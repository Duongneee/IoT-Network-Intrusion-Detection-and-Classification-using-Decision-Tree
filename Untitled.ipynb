```python
#Import thư viện 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
get_ipython().run_line_magic('matplotlib', 'inline')

# sklearn: data preprocessing
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder

# sklearn: train model
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, cross_validate, StratifiedKFold
from sklearn.metrics import precision_recall_curve, precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import roc_curve, auc, roc_auc_score, confusion_matrix, classification_report

# sklearn classifiers

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
#data set
initial_data = pd.read_csv('UNSW_NB15_training.csv')
```

```python
#hiển thị ra 10 mẫu đầu tiên của data set
initial_data.head(10)
```

```python
#Thông tin chi tiết của từng cột
initial_data.info()
```

```python
# làm sạch dữ liệu 
```

```python
#Kiểm tra các giá trị còn thiếu 
initial_data.isnull().sum()
```

```python
#Loại bỏ các hàng hoặc cột bị thiếu giá trị 
data_to_use = initial_data.dropna()
```

```python
#Dữ liệu sau khi được lọc
data_to_use.shape
```

```python
#Kiểm tra sự cân bằng trên kết quả.
```

```python
# Tách dữ liệu train thành tập dữ liệu và tập kết quả 
X = data_to_use.drop(axis=1, columns=['attack_cat']) # X is a dataframe
X = X.drop(axis=1, columns=['label'])

# Lấy cọto attack và label trong train
y1 = data_to_use['attack_cat'].values # y is an array
y2 = data_to_use['label'].values
```

```python
#Tính tỉ lệ
#Tính tỉ lệ số lần xuất hiện các giá trị khác nhau trong mảng Y2
def data_ratio(y2):
    '''
    Calculate Y2's ratio
    '''
    unique, count = np.unique(y2, return_counts=True)
    ratio = round(count[0]/count[1], 1)
    return f'{ratio}:1 ({count[0]}/{count[1]})'
```

```python
#In ra các tỉ lệ của y1 và y2 
y1_series = pd.Series(y1)
plt.figure(figsize=(13, 5))
sns.barplot(x=y1_series.value_counts().index, y=y1_series.value_counts(), label="Count")
plt.show()

y2_series = pd.Series(y2)
print('The class ratio for the original data:', data_ratio(y2_series))

# Đếm giá trị và sắp xếp lại các cột
count_data = y2_series.value_counts().sort_index()

# Vẽ biểu đồ đếm
plt.figure(figsize=(8, 5))
sns.barplot(x=count_data.index, y=count_data.values)
plt.ylabel('Count')

plt.show()
```

```python
#Nhập file để test
test_data = pd.read_csv('UNSW_NB15_testing-set.csv')
```

```python
#Tách tập thành các cột giá trị riêng và kết quả riêng test
X_test = test_data.drop(axis=1, columns=['attack_cat']) # X_test is a dataframe
X_test = X_test.drop(axis=1, columns=['label'])

#lấy các mảng giá trị của từng cột attack và label trong test
y1_test = test_data['attack_cat'].values # y is an array
y2_test = test_data['label'].values
```

```python
# Gán giá trị x, y1, y2 để phân biệt với tập test
X_train = X
y1_train = y1
y2_train = y2
```

```python
#Tách các cột có dtype là kiểu số và kiểu phân loại từ tập X
numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X_train.select_dtypes(include=['object', 'bool']).columns
```

```python
#Xem các cột là kiểu số
numerical_cols
```

```python
# Xem các cột là kiểu phân loại
categorical_cols
```

```python
# Định nghĩa các phương thức chuyển đổi cho các cột 
t = [('ohe', OneHotEncoder(drop='first'), categorical_cols), 
        ('scale', StandardScaler(), numerical_cols)] 

#tạo đối tượng với quy trình chuyển đổi t 
col_trans = ColumnTransformer(transformers=t)

# fit the transformation on training data
col_trans.fit(X_train)
    
# => col_trans sẽ học dữ liệu từ X_train sau đó áp dụng lên tập dữ liệu mới.
```

```python
#Chuyển đổi X_train
X_train_transform = col_trans.transform(X_train)
```

```python
#Chuyển đổi X_test
X_test_transform = col_trans.transform(X_test)
```

```python
#X_train sau khi chuyển
X_train_transform.shape
```

```python
#X_test sau khi chuyển
X_test_transform.shape
```

```python
# Xác định phương thức chuyển đổi LabelEncode() và khớp với y1_train
target_trans = LabelEncoder()
target_trans.fit(y1_train)
```

```python
# Áp dụng phương thức chuyển đổi trên y1_train và y1_test
y1_train_transform = target_trans.transform(y1_train)
y1_test_transform = target_trans.transform(y1_test)
```

```python
# nhãn y1-train sau khi chuyển đổi 
y1_train_transform
```

```python
#làm tương tự với y2
target_trans = LabelEncoder()
target_trans.fit(y2_train)
y2_train_transform = target_trans.transform(y2_train)
y2_test_transform = target_trans.transform(y2_test)
```

```python
#y2_train sau khi chuyển đổi 
y2_train_transform
```

```python
y2_test_transform
```

```python
#Mô hình hóa cây huấn luyện

#Xây dựng cây và huấn luyện 
DTclf = DecisionTreeClassifier()

# Train Decision Tree Classifer
DTclf = DTclf.fit(X_train_transform, y2_train_transform)

#Predict the response for test dataset
y_pred = DTclf.predict(X_test_transform)
```

```python
importances = DTclf.feature_importances_
indices = np.argsort(importances)
features = np.array(numerical_cols)
plt.title('Feature Importances of Decision Tree Classifier')
count = 10 # top # importance
plt.barh(range(count), importances[indices][len(indices)-count:], color='g', align='center')
plt.yticks(range(count), [features[i] for i in indices[len(indices)-count:]])
plt.xlabel('Relative Importance')
plt.show()
```

```python
fig = plt.figure(figsize=(35, 30))
DTtree = tree.plot_tree(DTclf, feature_names = numerical_cols.tolist(), class_names=['Normal', 'Attack'], fontsize=14, precision = 3, filled=True, rounded=True, max_depth= 5 )
for o in DTtree:
    arrow = o.arrow_patch
    if arrow is not None:
        arrow.set_edgecolor('red')
        arrow.set_linewidth(3)
```

```python
#Tính toán độ chính xác
from sklearn import metrics 
#Độ chính xác của mô hình
print("Accuracy:",metrics.accuracy_score(y2_test_transform, y_pred))

report=metrics.classification_report(y2_test_transform,y_pred)

DTclf_name=['Decision Tree Classifer','RegLog']

print('Reporting for %s:'%DTclf_name)

print(report)
```

```python
# trong đó :
# Precision = True Positives / (True Positives + False Positives)
# Recall = True Positives / (True Positives + False Negatives)
```
